```python
def display(img, title):
    img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.axis("off")
    plt.title(title)
    plt.show()
```

```python
def apply_grayscale(i_img):
    for y in range(i_img.shape[0]):
        for x in range(i_img.shape[1]):
            i_img[y,x]= 0.2989 * i_img[y,x,2] + 0.5871 * i_img[y,x,1] + 0.1140 * i_img[y,x,0]
    return i_img
```
